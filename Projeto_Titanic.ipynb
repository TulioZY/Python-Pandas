```python
import pandas as pd
from IPython.display import display

# Lê o conjunto de dados de passageiros do Titanic a partir de um arquivo CSV
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Exibe as primeiras 5 linhas do conjunto de dados
display(df.head())
```

```python
# Filtra os dados para incluir apenas mulheres com mais de 60 anos
df_filtrado = df[(df["Sex"] == "female") & (df["Age"] > 60)]

# Conta o número de passageiros neste subconjunto
num_mulheres_acima_60 = df_filtrado.shape[0]

# Exibe o resultado
display("Número de mulheres acima de 60:", num_mulheres_acima_60)
```

```python
# Filtra os dados para incluir apenas os passageiros que sobreviveram
df_filtrado2 = df[df["Survived"] == 1]

# Seleciona as colunas "Name", "Age" e "Survived"
df_selecionadas = df_filtrado2[["Name", "Age", "Survived"]]
    
# Exibe as informações selecionadas
display(df_selecionadas)

```

```python
# Mostra somente as pessoas com 35 anos de idade
display(df.loc[df["Age"] == 35])
```

```python
# Mostra somente as crianças entre 10 e 14 anos de idade
display(df.loc[(df["Age"] >= 10) & (df["Age"] <= 14)])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Conta o número de homens e mulheres que sobreviveram
num_MA_sobrev = df[(df["Sex"] == "male") & (df["Survived"] == 1)].shape[0]
num_FE_sobrev = df[(df["Sex"] == "female") & (df["Survived"] == 1)].shape[0]

# Define os dados a serem plotados
labels = ["Homens", "Mulheres"]
sizes = [num_MA_sobrev, num_FE_sobrev]

# Plot o gráfico de pizza
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.axis("equal")
plt.title("Sobreviventes por Gênero")
plt.show()
```

```python
# Calculando as porcentagens de sobreviventes baseado nas classes (algo como classe executiva X classe econômica)
total_classe1 = df.loc[df['Pclass'] == 1]['Survived'].count()
sobreviventes_classe1 = df.loc[(df['Pclass'] == 1) & (df['Survived'] == 1)]['Survived'].count()
porcentagem_classe1 = (sobreviventes_classe1 / total_classe1) * 100

total_classe2 = df.loc[df['Pclass'] == 2]['Survived'].count()
sobreviventes_classe2 = df.loc[(df['Pclass'] == 2) & (df['Survived'] == 1)]['Survived'].count()
porcentagem_classe2 = (sobreviventes_classe2 / total_classe2) * 100

total_classe3 = df.loc[df['Pclass'] == 3]['Survived'].count()
sobreviventes_classe3 = df.loc[(df['Pclass'] == 3) & (df['Survived'] == 1)]['Survived'].count()
porcentagem_classe3 = (sobreviventes_classe3 / total_classe3) * 100

display("Porcentagem de sobreviventes na classe 1: {:.2f}%".format(porcentagem_classe1))
display("Porcentagem de sobreviventes na classe 2: {:.2f}%".format(porcentagem_classe2))
display("Porcentagem de sobreviventes na classe 3: {:.2f}%".format(porcentagem_classe3))
```

```python
# Gerando gráfico baseado nos resultados do código acima
classe = ['Classe 1', 'Classe 2', 'Classe 3']
porcentagens = [porcentagem_classe1, porcentagem_classe2, porcentagem_classe3]

plt.bar(classe, porcentagens)
plt.xlabel('Classe')
plt.ylabel('Porcentagem de Sobreviventes (%)')
plt.title('Porcentagem de Sobreviventes por Classe')
plt.show()
```
